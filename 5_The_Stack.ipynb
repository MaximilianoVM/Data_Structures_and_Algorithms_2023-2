<h3>The call stack</h3> 

```python

def greet2(name):
    print ("how are you, " + name + "?")
def bye():
    print ("ok bye!")

def greet(name):
    print ("hello, " + name + "!")
    greet2(name)
    print ("getting ready to say bye...")
    bye()
```

```python
greet('max')
```

<h2>The call stack with recursion</h2>

```python
def fact(x):
    if x == 1:
        return 1
    else:
        return x * fact(x-1)

fact(3)
```
