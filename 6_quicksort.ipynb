<h3><b>2.</b> Write a recursive function to count the number of items in a list.</h3>

```python
def n_items(lis):
    if len(lis) == 0: return 0
    else:
        return 1 + n_items(lis[1:])
```

<h3><b>3.</b> Write a recursive function to find the maximum number in a list.</h3>

```python
def find_max(lst):
    # Base case
    if len(lst) == 0: return None
    elif len(lst) == 1:
        return lst[0]
    
    # Recursive case 
    else:
        # Divide
        mid = len(lst) // 2
        left_half = lst[:mid]
        right_half = lst[mid:]

        # Recursion in each half
        max_left = find_max(left_half)
        max_right = find_max(right_half)

        #return the larger one
        return max(max_left, max_right)


prueba = [3, 8, 1, 6, 4, 2, 9, 7]
maximum = find_max(prueba)
print("Max:", maximum)
```

<h3><b>4.</b> Binary search is a divide-and-conquer algorithm. Can you come up with the base
case and recursive case for binary search?

```python
def binary_search(arr, target, left, right):
    if left > right:
        return -1  # Base case: Element not found

    mid = (left + right) // 2
    if arr[mid] == target:
        return mid  # Base case: Element found at index mid
    elif arr[mid] > target:
        return binary_search(arr, target, left, mid - 1)  # Recursive case: Search in the left half
    else:
        return binary_search(arr, target, mid + 1, right)  # Recursive case: Search in the right half
```

<h3><b>5.</b> Try the quicksort algorithm, describe its functionality under different scenarios(i.e. 20 elements, negative elements, etc.)

```python
def quicksort_1(arr):
    if len(arr) <= 1:
        return arr  # Base case: An array with 0 or 1 element is already sorted

    pivot = arr[len(arr) // 2]  # Choose a pivot element (often the middle one)
    left = [x for x in arr if x < pivot]  # Elements less than the pivot
    middle = [x for x in arr if x == pivot]  # Elements equal to the pivot
    right = [x for x in arr if x > pivot]  # Elements greater than the pivot

    return quicksort_1(left) + middle + quicksort_1(right)  # Recursively sort the sub-arrays

# Example usage:
my_array = [3, 6, 8, 10, 1, 2, 1]
sorted_array = quicksort_1(my_array)
print(sorted_array)  # Output: [1, 1, 2, 3, 6, 8, 10]
```
