```python
import matplotlib.pyplot as plt
import math
```

```python
# Funci칩n para calcular la distancia euclidiana entre dos puntos (x, y)
def calcular_distancia(punto1, punto2):
    return math.sqrt((punto1[0] - punto2[0])**2 + (punto1[1] - punto2[1])**2)
```

```python
# Funci칩n para encontrar el vecino m치s cercano a un punto en una lista de puntos
def vecino_mas_cercano(punto, lista_puntos):
    distancia_minima = float('inf')
    vecino_cercano = None

    for vecino in lista_puntos:
        distancia = calcular_distancia(punto, vecino)
        if distancia < distancia_minima:
            distancia_minima = distancia
            vecino_cercano = vecino

    return vecino_cercano
```

```python
# Datos de prueba
puntos = [(1, 2), (5, 6), (3, 4), (9, 8), (2, 7)]

# Punto de prueba
punto_prueba = (4, 5)
```

```python
# Encontrar el vecino m치s cercano al punto de prueba
vecino_cercano = vecino_mas_cercano(punto_prueba, puntos)
print(vecino_cercano)
```

```python
# Visualizaci칩n de los puntos y el vecino m치s cercano
xs, ys = zip(*puntos)
plt.scatter(xs, ys, marker='o', label='Puntos')
plt.scatter(*punto_prueba, color='red', marker='x', label='Punto de prueba')
plt.scatter(*vecino_cercano, color='green', marker='o', s=100, label='Vecino m치s cercano')

plt.title('Vecinos m치s cercanos')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
